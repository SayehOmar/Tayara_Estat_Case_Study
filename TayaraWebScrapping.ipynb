```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd 
```

Function to extract the HTML from a the web  page


```python
def extract(page):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36 OPR/101.0.0.0"
    }
    url = f"https://www.tayara.tn/ads/c/Immobilier/?page={page}"  
    r = requests.get(url, headers=headers)  
    soup = BeautifulSoup(r.content, "html.parser")
    return soup
```

Extracting the data (Locations,Prices,Estat_types,Description) from the HTML 

```python
Locations=[]
Prices=[]
Estat_types=[]
Descriptions=[]

for i in range (300):
    div_elements = extract(i+1).find_all('div', class_='pb-2 px-3 z-10 flex-none -mt-1')
   
    if div_elements:
        for div in div_elements:
            text_parts = []
            # Extract text from specific tags within the div
            price = div.find('data').text.strip() if div.find('data') else ""
            description = div.find('span',class_='truncate text-3xs md:text-xs lg:text-xs w-3/5 font-medium text-neutral-500').text.strip() if div.find('span') else ""
            Estat_type = div.find('h2').text.strip() if div.find('h2') else ""
            Location= div.find('span',class_='line-clamp-1 truncate text-3xs md:text-xs lg:text-xs w-3/5 font-medium text-neutral-500').text.strip()if div.find('span')else""
            
            # Append text parts to the list
            if price:
                text_parts.append(price)
            if description:
                text_parts.append(description)
            if Estat_type:
                text_parts.append(Estat_type)
            if Location:
                text_parts.append(Location)
            
            # Join and print the text parts
            combined_text = " / ".join(text_parts)
            print(combined_text)
    else:
        print("No <div> elements found in the HTML.")

    string = combined_text.split('/')
    Prices.append(string[0].strip())
    Estat_types.append(string[1].strip())
    Descriptions.append(string[2].strip())
    Locations.append(string[3].strip())
```

Joining the data (Locations,Prices,Estat_types,Description) together within a DataFrame 

```python

df=pd.DataFrame({'Prices':Prices,'Estat_types':Estat_types, 'Descriptions':Descriptions,'Locations':Locations})
print (df)

```

Exporting the DataFrame into a csv file

```python
df.to_csv('TayraWebScrapping.csv',index=False)
```
